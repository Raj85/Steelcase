
# Pieces in Network file - Combining daily files

*   Loop through all the files from a folder and combine them into one cvs file





# Global Settings and Imports

```python
import os
import pandas as pd
import sqlite3
from glob import glob
import time
```

### Fucntion to iterate over all the raw .xslm files

```python
# Create an empty DataFrame to store the combined data
combined_df = pd.DataFrame()

# Loop over all files in the directory
for file in glob('*.xlsm'):
    start_time = time.time()
    # Read the Excel file into a DataFrame
    df = pd.read_excel(file, sheet_name='DATA' )
    # Add a new column with the file name (without the extension)
    df['report_pull_date'] = file[:8]
    # concatenate the DataFrame to the combined DataFrame
    combined_df = pd.concat([combined_df, df], ignore_index=True)
    end_time = time.time()
    total_time =round((end_time - start_time)/60,2)    
    print(f"Time taken for iteration {file}: {total_time} mins")
```

### Save the file into a .csv and .parquet file

```python
# Save the DataFrame as a pipe-delimited CSV file with a text qualifier
combined_df.to_csv('six_month_network_data.csv', sep='|', quotechar='"', index=False)
combined_df.to_parquet('new_df')
```
